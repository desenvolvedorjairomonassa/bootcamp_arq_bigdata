```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/desenvolvedorjairomonassa/bootcamp_arq_bigdata/main/modulo2/dados_jogadores-1.csv',
                 sep=';',encoding='latin-1')
```

```python
#verificar numeros nulos
df.isnull().sum()
```

```python
# numero de total de partidas nulas
df.loc[df['total_partidas'].isnull()]
```

```python
# alguns totais de partidas estÃ£o nulas
def total (vitorias, derrotas):
  return vitorias+derrotas

df.loc[df['total_partidas'].isnull(),'total_partidas'] =  df.apply(lambda x: total(x['num_vitorias'],x['num_derrotas']), axis=1)
```

```python
df.describe()
```

```python
#verificar num_vitorias nulas
df.loc[df['num_vitorias'].isnull()]
```

```python
# numero de vitorias nulas atualizar com total - derrotas
def vitorias (total, derrotas):
  return total-derrotas

df.loc[df['num_vitorias'].isnull(),'num_vitorias'] = df.apply(lambda x :  vitorias(x['total_partidas'], x['num_derrotas']) , axis=1  )
```

```python
#verificar num_derrotas nulas
df.loc[df['num_derrotas'].isnull()]
```

```python
def num_derrotas (vitorias, total):
  return total-vitorias

df.loc[df['num_derrotas'].isnull(),'num_derrotas'] = df.apply(lambda x :  num_derrotas(x['num_vitorias'], x['total_partidas']) , axis=1  )
```

```python
#Qual o paÃ­s possui a menor mÃ©dia de vitÃ³rias? No geral (independente do jogo). 
df.groupby('pais')['pais','num_vitorias'].sum().sort_values(ascending=True, by='num_vitorias').head(4)
```

```python
#verificar numeros nulos
df.isnull().sum()
```

```python
#Qual Ã© a soma de derrotas do Brasil no jogo Dama? 
df.loc[(df['pais']=='Brasil')&(df['jogo']=='Dama')]['num_derrotas'].sum()
```

```python
#Qual o jogo possui o maior nÃºmero de partidas? 
df.groupby('jogo')['jogo','total_partidas'].sum().sort_values(ascending=False, by='total_partidas').head(1)
```

```python
#Qual o jogo de menor preferÃªncia entre os homens? 
df.loc[df['genero']=='Masculino'].groupby('jogo')['jogo','total_partidas'].sum().sort_values(ascending=True, by='total_partidas').head()
```
