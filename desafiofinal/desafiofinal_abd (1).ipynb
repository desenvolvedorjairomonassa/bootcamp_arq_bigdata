```python
#Quantos produtos distintos existem no dataset de mercado?
# pegando um conjunto , pois não possue repetido, usando o set
conjunto_final = set()
mercado = nutricional.fillna(0)
for i in mercado.columns[1:]:
  #print(mercado[i].unique())
  conjunto=set(mercado[i].unique())
  conjunto_final=conjunto_final.union(conjunto)
#remover o nulo
conjunto_final.remove(0)  #que foi subsitituido por todos os Nan
#tamanho
print(len(conjunto_final))
```

```python
#montar um dataframe com todos os produtos unicos e mais o id na primeira posição
pivot_mercado = nutricional.iloc[:,0:1]
for i in conjunto_final:
  pivot_mercado[i] = 0 

#preencher a matriz com os produtos do usuario
#pivot_mercado = pivot_mercado.apply(lambda row: rowIndex , axis=1)
for idx,row in pivot_mercado.iterrows():
  lista = nutricional.iloc[idx,1:].reset_index(drop=True)
  lista.dropna(inplace=True) 
  for item in lista:
    pivot_mercado[item][idx] = 1

```
